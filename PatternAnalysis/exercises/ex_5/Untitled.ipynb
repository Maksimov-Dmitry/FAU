```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm, rcParams
%matplotlib inline
from sklearn.datasets import make_blobs
scatter_params = {'s': 15, 'alpha': .7}
plt.style.use('seaborn')
```

```python
n_samples = 1000
random_seed = 235
X, y = make_blobs(n_samples=n_samples, centers=4, cluster_std=[1.5, 4, 3, 1], random_state=random_seed)
```

```python
def plot(X, y):
    cmap = cm.get_cmap("Spectral")
    colors = {0: "red", 1: "green", 2: 'blue', 3:'lightblue', 4:'yellow', 5:'pink', 6:'orange', 7:'purple', 8:'magenta'}
    plt.scatter(X[:, 0], X[:, 1], c=[colors[i] for i in y], **scatter_params)
    plt.title(f'number of clusters: {len(np.unique(y))}')
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
```

```python
plot(X, y)
```

```python
data = np.random.uniform(low=X.min(axis=0), high=X.max(axis=0), size=(n_samples,2))
plt.scatter(data[:, 0], data[:, 1], **scatter_params)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
