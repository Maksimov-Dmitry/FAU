```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm, rcParams
%matplotlib inline
from sklearn.datasets import make_blobs
scatter_params = {'s': 15, 'alpha': .7}
rcParams['figure.figsize'] = (12, 10)
from sklearn.cluster import MeanShift, KMeans
```

```python
n_samples = 1000
random_seed = 235
X, y = make_blobs(n_samples=n_samples, centers=3, cluster_std=[1.5, 4, 3], random_state=random_seed)
```

```python
def plot(X, y):
    cmap = cm.get_cmap("Spectral")
    plt.scatter(X[:, 0], X[:, 1], c=[cmap(float(i) /10) for i in y], **scatter_params)
    plt.title(f'number of clusters: {len(np.unique(y))}')
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
```

```python
meanshift = MeanShift(bandwidth=2).fit(X)
```

```python
y_meanshift = meanshift.predict(X)
```

```python
plot(X, y_meanshift)
```

```python
plot(X, y)
```
